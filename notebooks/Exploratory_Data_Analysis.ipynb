```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="loan_status",
    hue = 'loan_status',
    data=df,
    palette={"Approved": "green", "Rejected": "red"},
legend = False
)

plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
```

```python
df["loan_status"].value_counts(normalize=True) * 100
```

##  Loan Status Distribution Analysis

The distribution of the target variable **loan_status** shows the following:

- **Approved:** 62.215%  
- **Rejected:** 37.784%

###  Observations:
- The dataset exhibits **moderate class imbalance**, with a higher proportion of approved loans.
- Approximately two-thirds of the loan applications are approved.
- There is a difference of about 25% between approved and rejected cases.

###  Business Interpretation:
This suggests that the bank approves a majority of loan applications, which may indicate:
- Strong applicant financial profiles
- Favorable credit evaluation criteria
- Risk assessment policies allowing higher approval rates

###  Impact on Machine Learning Model:
- The imbalance may cause the model to favor the majority class (Approved).
- Accuracy alone may not be sufficient for evaluation.
- It is recommended to evaluate using:
  - Precision
  - Recall
  - F1-Score
  - ROC-AUC
- If required, class balancing techniques such as `class_weight="balanced"` or SMOTE can be applied.

Overall, while the imbalance is not extreme, it should be considered during model training and evaluation.

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x="loan_status",
    y="cibil_score",
    data=df
)

plt.title("CIBIL Score vs Loan Status")
plt.show()
```

```python
df.groupby("loan_status")["cibil_score"].mean()
```

```python
df.groupby("loan_status")["cibil_score"].median()
```

```python
sns.histplot(data=df, x="cibil_score", hue="loan_status", kde=True)
plt.title("CIBIL Score Distribution by Loan Status")
plt.show()
```

##  CIBIL Score vs Loan Status Analysis

The distribution and boxplot analysis of CIBIL score shows a clear separation between approved and rejected loan applications.

###  Key Observations:

- Approved applicants generally have CIBIL scores above 600.
- Rejected applicants mostly have CIBIL scores between 300 and 500.
- The median CIBIL score of approved loans is significantly higher than that of rejected loans.
- There is minimal overlap between the two distributions.

###  Business Interpretation:

CIBIL score strongly influences loan approval decisions. Higher credit scores indicate better repayment history and lower financial risk, making applicants more likely to receive approval.

###  Conclusion:

CIBIL score is the most impactful feature in predicting loan approval in this dataset. It is expected to have high feature importance in machine learning models.


```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x="income_annum",
    y="loan_amount",
    data=df
)

plt.title("Income vs Loan Amount")
plt.xlabel("Annual Income")
plt.ylabel("Loan Amount")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x="income_annum",
    y="loan_amount",
    hue="loan_status",
    data=df
)

plt.title("Income vs Loan Amount by Loan Status")
plt.show()
```

```python
df[["income_annum", "loan_amount"]].corr()
```

##  Income vs Loan Amount Analysis

The scatter plot shows a strong positive linear relationship between annual income and loan amount.

###  Key Observations:

- Loan amount increases proportionally with income.
- There is a clear linear trend indicating strong correlation.
- Approved and rejected applications are distributed across similar income and loan ranges.
- Income alone does not clearly distinguish between approved and rejected loans.

###  Business Interpretation:

Banks tend to offer higher loan amounts to applicants with higher income, as repayment capacity is directly linked to earnings. However, income alone is not sufficient to determine loan approval status.

###  Conclusion:

While income_annum and loan_amount are strongly correlated, loan approval decisions likely depend more heavily on credit risk indicators such as CIBIL score rather than income alone.


```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

##  Correlation Heatmap Analysis

The correlation heatmap reveals strong relationships among financial features in the dataset.

###  Key Observations:

- income_annum and loan_amount show very strong positive correlation (0.93).
- Asset-related features (residential, commercial, luxury, and bank assets) are highly correlated with income.
- luxury_assets_value also shows strong correlation with loan_amount.
- CIBIL score shows near-zero correlation with income and asset values.

###  Multicollinearity:

There is high multicollinearity among asset-related features. For modeling purposes, these may need to be combined (e.g., total_assets) or reduced to avoid redundancy.

###  Business Interpretation:

Income determines loan size, while CIBIL score (creditworthiness) independently influences approval decisions. This suggests that repayment capacity and credit history are evaluated separately in the approval process.

###  Conclusion:

CIBIL score is likely the most important independent feature for predicting loan approval, while income and asset values primarily influence loan amount.


## Final Insights from EDA

1. CIBIL score is the strongest predictor of loan approval.
2. Income and asset values strongly influence loan amount.
3. Moderate class imbalance exists in loan_status.
4. Loan amount is right-skewed and may require log transformation.
5. Multicollinearity exists among asset-related features.

These insights will guide preprocessing and model selection.
